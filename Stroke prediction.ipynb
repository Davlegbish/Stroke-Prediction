# Feature Enginnering (Dummies)

```python
stroke.select_dtypes(['object']).columns
```

```python
dummies=pd.get_dummies(stroke[['gender','ever_married','work_type','Residence_type','smoking_status']],drop_first=True)
stroke=pd.concat([stroke.drop(['gender','ever_married','work_type','Residence_type','smoking_status'],axis=1),dummies],axis=1)
```

```python
stroke.head()
```

# Model Creation

```python
X=stroke.drop('stroke',axis=1).values
y=stroke['stroke'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',patience=25,verbose=1,)
```

```python
X_train.shape
```

```python
model=Sequential()
#frist layer
model.add(Dense(units=14,activation='relu'))
model.add(Dropout(0.2))
#hidden layer
model.add(Dense(units=10,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=6,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=4,activation='relu'))
model.add(Dropout(0.2))
#last Layer
model.add(Dense(units=1,activation='sigmoid'))
#compile model
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=700,verbose=1,callbacks=[early_stop])
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

# Model Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predict=np.argmax(model.predict(X_test),axis=1)
```

```python
print(classification_report(y_test,predict))
```

```python
print(confusion_matrix(y_test,predict))
```
